```python
df['frag_key'] = (
    df['FragmentType'].astype(str) + "_" +
    df['FragmentNumber'].astype(str) + "_" +
    df['FragmentLossType'].astype(str) + "_ch" +
    df['FragmentCharge'].astype(str)
)

df[['FragmentType','FragmentNumber','FragmentLossType','FragmentCharge','frag_key']].head(10)
```

```python
duplicate_check = (
    df.groupby(['SpectrumID','frag_key'])
      .size()
      .reset_index(name='count')
)

duplicate_check[duplicate_check['count'] > 1]
```

```python
agg = (
    df.groupby(['SpectrumID', 'frag_key'])['RelativeIntensity']
      .mean()
      .reset_index()
)

agg.head(10)
```

```python
check_after = (
    agg.groupby(['SpectrumID','frag_key'])
       .size()
       .reset_index(name='count')
)

check_after[check_after['count'] > 1]
```

```python
max_per_spec = agg.groupby('SpectrumID')['RelativeIntensity'].max()

max_per_spec

```

```python
agg['RelNorm'] = agg['RelativeIntensity']
agg.head()
```

```python
mat = agg.pivot(
    index='SpectrumID',
    columns='frag_key',
    values='RelNorm'
).fillna(0)

mat.head(10)
```

```python
print("Matrix shape:", mat.shape)
```

```python
mat.iloc[1]
```

```python
from sklearn.metrics.pairwise import cosine_similarity

sims = cosine_similarity(mat.values)

sim_df = pd.DataFrame(
    sims,
    index=mat.index,
    columns=mat.index
)

sim_df
```

```python
# Bỏ đường chéo
mask = ~np.eye(sim_df.shape[0], dtype=bool)
min_similarity = sim_df.where(mask).min().min()

min_similarity
```
